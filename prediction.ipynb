```python
import pickle
```

```python
country_to_model = pickle.load(open('country_to_model.pickle','rb'))
normalization_dict = pickle.load(open('feature_normalization_dict.pickle','rb'))
```

```python
import pandas as pd
pred_data = pd.read_csv('final_test_data.csv')
```

```python
pred_data.head()
```

```python
def norm_row(row):
    row.to_dict()
    inp = []
    for f in ['Year','GDP','Trade','Inflation','Exports']:
        inp.append((row[f]-normalization_dict[f]['min'])/(normalization_dict[f]['max']-normalization_dict[f]['min']))
    return inp
```

```python
final_output = []
for idx,row in pred_data.iterrows():
    inp = norm_row(row)
    prediction = country_to_model[row['Country Code']].predict(inp)[0]
    line = [row['Country Name'],row['Country Code'],row['Year'],row['Inflation'],row['GDP'],row['Exports'],row['Trade'],prediction]
    final_output.append(line)
```

```python
final_output = pd.DataFrame(final_output,columns=['Country Name','Country Code','Year','Inflation','GDP','Exports','Trade','Balance'],index=None)
```
