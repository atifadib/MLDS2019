```python
from sklearn.ensemble import GradientBoostingRegressor
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_excel('normed_train_data.xlsx')
```

```python
train_data.head()
```

```python
from tqdm import tqdm
country_to_model = {}
for country in tqdm(set(train_data['Country Code'])):
    gbr = GradientBoostingRegressor(loss ='huber', max_depth=10)
    ctrain_data = train_data[train_data['Country Code']==country]
    X = ctrain_data[['norm_year','norm_gdp','norm_trade','norm_inflation','norm_exports']].values
    Y = ctrain_data[['Balance']].values
    gbr.fit(X,Y)
    country_to_model[country]=gbr
```
